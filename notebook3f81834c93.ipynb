# Education and Literacy Analysis in Nepal (Census 2021)

## Introduction

This project analyzes literacy patterns in Nepal using census data. 
The dataset represents population aged 5 years and above.

## load data

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/rabinbro/nepal-literacy-data-2021/formated_nr_indv_table13_literacystatus.csv")

df.head()
```

## Data cleaning

```python
df.columns
```

```python
df.info()
```

```python

df = df.iloc[2:]

df = df.reset_index(drop=True)

df.head()
```

```python
df
```

```python
df.columns
```

```python
df.columns = ["Area", "Sex", "Literate", "Illiterate", "Not_Stated", "Total", "Extra"]
```

```python
df
```

```python
df = df.drop(columns=["Extra"])
```

```python
df
```

```python
df.info()
```

***changing the object into string***

```python
df["Literate"] = pd.to_numeric(df["Literate"], errors='coerce')
df["Illiterate"] = pd.to_numeric(df["Illiterate"], errors='coerce')
df["Total"] = pd.to_numeric(df["Total"], errors='coerce')
```

```python
df.info()
```

removing the missing value 

```python
df.isnull().sum()
```

ohhh , no value is missing 

## National Literacy Overview

***Literacy by gender***

```python
df.groupby("Sex")[["Literate", "Illiterate"]].sum()
```

```python
nepal=df[df["Area"].str.contains("Nepal", na=False)]
```

```python
total_literate = nepal["Literate"].sum()
total_illiterate = nepal["Illiterate"].sum()

print("Total Literate:", total_literate)
print("Total Illiterate:", total_illiterate)
```

```python
nepal
```

***Literacy vs Illiteracy***

```python
import matplotlib.pyplot as plt

labels = ["Literate", "Illiterate"]
values = [total_literate, total_illiterate]

plt.bar(labels, values)
plt.title("Literacy Status in Nepal (Age 5+)")
plt.ylabel("Population")
plt.show()
```

***Gender Comparison***

```python
nepal.groupby("Sex")[["Literate", "Illiterate"]].sum().plot(kind="bar")

plt.title("Literacy by Gender in Nepal")
plt.ylabel("Population")
plt.show()
```

***lets look for provincial***

## Province Analysis

```python
df["Area"].unique()
```

```python
provinces = df[
    df["Area"].isin([
        "Koshi", "Madhesh", "Bagmati",
        "Gandaki", "Lumbini", "Karnali", "Sudur Paschim"
    ])
]
```

```python
provinces = provinces[provinces["Sex"] == "Total"]
```

```python
provinces
```

```python
import matplotlib.pyplot as plt

provinces.plot(x="Area", y=["Literate", "Illiterate"], kind="bar")

plt.title("Literacy vs Illiteracy by Province in Nepal")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```

***some point to be noted***
1.Bagmati Province has the highest literacy level.
2.Some provinces still have high illiteracy rates.
3.There is a noticeable gap between literate and illiterate populations across provinces.
4.Education development is uneven across regions.

## District Analysis

```python
districts = df[
    ~df["Area"].str.contains(
        "Nepal|Mountain|Hill|Tarai|Koshi|Madhesh|Bagmati|Gandaki|Lumbini|Karnali|Sudur Paschim",
        na=False
    )
]
```

```python
districts = districts[districts["Sex"] == "Total"]
```

```python
districts
```

oh ,shit we have urban municipalities

```python
districts = df[
    ~df["Area"].str.contains(
        "Nepal|Mountain|Hill|Tarai|Koshi|Madhesh|Bagmati|Gandaki|Lumbini|Karnali|Sudur Paschim|Urban Municipalities|Rural Municipalities",
        na=False
    )
]
```

```python
districts
```

oh we have some duplicated district here 

```python
districts = districts[districts["Sex"] == "Total"]
```

```python
districts 
```

```python
districts.head(10)
```

```python
districts.tail
```

```python
top_literate = districts.sort_values(by="Literate", ascending=False)

top_literate.head(10)
```

sabi bhanda badii schixit yuwa pusta

```python
import matplotlib.pyplot as plt

top_literate.head(10).plot(x="Area", y="Literate", kind="bar")

plt.title("Top 10 Districts by Literacy in Nepal")
plt.xticks(rotation=45)
plt.show()
```

```python
top_illiterate = districts.sort_values(by="Illiterate", ascending=False)

top_illiterate.head(10)
```

```python
import matplotlib.pyplot as plt

top_literate.tail(10).plot(x="Area", y="Illiterate", kind="bar")

plt.title("Top 10 Districts by Illiterate in Nepal")
plt.xticks(rotation=45)
plt.show()
```

***Districts like Kathmandu and Lalitpur have the highest literacy levels.
Remote districts show lower literacy.
There is a strong inequality in education across districts.***
